```python
#mount to google drive because dataset is in the google drive

from google.colab import drive
drive.mount('/content/drive')
```

```python
# read dataset

df = pd.read_csv('/content/drive/MyDrive/Data/archive/Translate.csv')
df.head()
```

```python
#dataset setails

df.describe()
```

```python
#read dataset from csv file

title_string=df.review_title.values
print(title_string)
```

```python
title_sequence=list(title_string)
print(title_sequence)
```

```python
#huggingface model used

pipe = pipeline(task='text2text-generation', model='facebook/m2m100_418M')
```

```python
#convert to sequence

trans_ms = pipe(title_sequence, forced_bos_token_id=pipe.tokenizer.get_lang_id(lang='en'))
```

```python
print(trans_ms)
```

```python
#convert to df

translated = pd.DataFrame(trans_ms)
```

```python
df['trans_ms']=translated
df.head()
```

```python
#save as csv

df.to_csv('review_title.csv')
```
